```python
cars_df = pd.read_csv("Cars.csv")
cars_df.head(10)
```

```python
cars_df = cars_df[['VOL' ,'MPG']]
```

```python
cars_df
```

```python
cars_df.set_index('VOL')
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
plt.plot(cars_df.VOL);
plt.xlabel('MPG');
plt.ylabel('vol');
```

```python
print('Mean: ', round(cars_df.VOL.mean(), 4))
print('Median: ', round(cars_df.VOL.median(), 4))
print('Mode: ', round(cars_df.VOL.mode(), 4))
print('Variance: ', round(cars_df.VOL.var(), 4))
print('Standard Deviation: ', round(cars_df.VOL.std(), 4))
```
